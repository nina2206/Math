```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
#преобразует полярные координаты в декартовы
R = 3
corner = np.pi/2
decX = R*np.cos(corner)
decY = R*np.sin(corner)
print(decX)
print(decY)
```

```python
# график окружности в полярных координатах
x = np.linspace(0, 5, 100)
R = 5
y = np.sqrt(R**2-x**2)
plt.polar(x, y)
```

```python
# график прямой в полярных координатах
x = np.linspace(0, 5, 100)
y = x+2
plt.polar(x, y)
```
