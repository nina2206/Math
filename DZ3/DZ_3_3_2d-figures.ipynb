```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
x = []
y_rou1 = []
y_rou2 = []
r = 10
for i in range (-1000, 1000):
    x_i = i/50
    x.append(x_i)
    y_rou1.append(np.sqrt(r**2 - x_i**2))
    y_rou2.append(-np.sqrt(r**2 - x_i**2))
plt.plot(x, y_rou1, color='red')
plt.plot(x, y_rou2, color='red')
plt.xlabel("x")
plt.ylabel("y")
```

```python
x = []
y_elp1 = []
y_elp2 = []
el_1 = 4
el_2 = 2
for i in range (-1000, 1000):
    x_i = i/100
    x.append(x_i)
    y_elp1.append(np.sqrt(el_1**2-(x_i*el_1/el_2)**2))
    y_elp2.append(-np.sqrt(el_1**2-(x_i*el_1/el_2)**2))    
plt.plot(x, y_elp1, color='red')
plt.plot(x, y_elp2, color='red')
plt.xlabel("x")
plt.ylabel("y")
```

```python
x = []
y_gip1 = []
y_gip2 = []
g_1 = 12
g_2 = 2
for i in range (-1000, 1000):
    x_i = i/120
    x.append(x_i)
    y_gip1.append(np.sqrt(g_1**2+(g_1*x_i/g_2)**2))
    y_gip2.append(-np.sqrt(g_1**2+(g_1*x_i/g_2)**2))    
plt.plot(x, y_gip1, color='red')
plt.plot(x, y_gip2, color='red')
plt.xlabel("x")
plt.ylabel("y")
```
