```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-3, 5, 500)
y1 = x**2-1
y2 = (np.exp(x)-1)/x+1
plt.plot(x, y1, label='y=x^2-1', c="orange")
plt.plot(x, y2, label='y=(exp(x)-1)/x+1', c='blue')
plt.ylim(-3, 10)
plt.grid(True)
plt.legend(loc='upper left')
plt.show()
print('точки пересечения графиков лежат в диапазонах x: (-2, -1), (2, 3)')
```

```python
from scipy.optimize import fsolve
def f(p):
    x = p
    return (np.exp(x)-x**3+2*x-1)
print("Уравнения системы принимают нулевые значения при значениях х:")
x1 = fsolve(f, -2)
print("%.2f"%x1)
x2 = fsolve(f, 0)
print("%.2f"%x2)
x3 = fsolve(f, 2)
print("%.2f"%x3)
x4 = fsolve(f, 4)
print("%.2f"%x4)
print("Значения функции по условиям системы должны быть больше нуля.")
print(f" Функция принимает положительные значения в диапазонах: (-{np.inf};{x1}), ({x2}, {x3}), ({x4}, {np.inf})")
x = np.linspace(-3, 10, 500)
plt.plot([-5, 10], [0, 0], c="red", linewidth=0.8, linestyle='--')
y3 = np.exp(x)-x**3+2*x-1
plt.plot(x, y3, label='y=(exp(x)-x^3+2*x-1', c='green')
plt.ylim(-5, 10)
plt.grid(True)
plt.legend(loc='upper left')
plt.show()
```
