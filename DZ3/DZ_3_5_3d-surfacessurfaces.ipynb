```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = figure()
ax = Axes3D(fig)
X1 = np.arange(-5, 5, 0.5)
Y1 = np.arange(-5, 5, 0.5)
X1, Y1 = np.meshgrid (X1, Y1)
Z1 = 2*X1 + 2*Y1
ax.plot_wireframe (X1, Y1, Z1, color='purple')
X2 = np.arange(-5, 5, 0.5)
Y2 = np.arange(-5, 5, 0.5)
X2, Y2 = np.meshgrid (X2, Y2)
Z2 = 2*X2 + 2*Y2+50
ax.plot_wireframe(X2, Y2, Z2, color='orange')
show()
```

```python
fig = figure()
ax = Axes3D(fig)
X1 = np.arange(-5, 5, 0.5)
Y1 = np.arange(-5, 5, 0.5)
X1, Y1 = np.meshgrid (X1, Y1)
Z1 = X1**2 + Y1**2
ax.plot_wireframe (X1, Y1, Z1, color='purple')
X2 = np.arange(-5, 5, 0.5)
Y2 = np.arange(-5, 5, 0.5)
X2, Y2 = np.meshgrid (X2, Y2)
Z2 = X2**2 - Y2**2 - 30
ax.plot_wireframe(X2, Y2, Z2, color='orange')
show()
```
