```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10, 10, 200)
plt.plot(x, -(np.sqrt(2*x+6))+1, 'red', linewidth=1)
plt.plot(x, +(np.sqrt(2*x+6))+1, 'red', linewidth=1)
plt.grid(True)
print('17.6.5. Уравнение параболы')
plt.show()
```

```python
x = np.linspace(-10, 10, 200)
plt.plot(x, -np.sqrt(5-3/5*(2+x)**2)+3, 'red', linewidth=1)
plt.plot(x, +np.sqrt(5-3/5*(2+x)**2)+3, 'red', linewidth=1)
plt.grid(True)
print('17.6.6. Уравнение эллипса')
plt.show()
```

```python
x = np.linspace(-10, 10, 200)
plt.plot(x, -np.sqrt(2*x**2-16)+3, 'red', linewidth=1)
plt.plot(x, +np.sqrt(2*x**2-16)+3, 'red', linewidth=1)
plt.grid(True)
print('17.6.7. Уравнение гиперболы')
plt.show()
```

```python
x = np.linspace(-100, 100, 1000)
plt.plot(x, -np.sqrt(2/3*(x-7)**2-100)+7, 'red', linewidth=1)
plt.plot(x, +np.sqrt(2/3*(x-7)**2-100)+7, 'red', linewidth=1)
plt.grid(True)
print('17.6.8. Уравнение гиперболы')
plt.show()
```
