```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
# график функции при конкретном значении параметра
x = np.linspace(100, 500, 1000)
a = 0.15
y = np.sin(0.15*x)
plt.plot(x, y, label='y = sin(x)', c="orange")
plt.plot([0, 500], [0, 0], c="red", linewidth=0.8, linestyle='--')
plt.ylim(-2, 2)
plt.grid(True)
plt.legend(loc='upper left')
plt.show()
```

```python
#взаимозависимость параметров в заданных диапазонах
a = np.linspace(0.01, 0.02, 500)
x = np.pi/a
print('для всех а от 0,01 до 0,02 и  от 100 до 500 множество решений предствалено функцией x=pi/a')
plt.plot(a, x, label='x=pi/a')
plt.ylim(0, 600)
plt.grid(True)
plt.legend(loc='upper left')
plt.show()
print('например: для a = 0.15 x равен: ', np.pi/0.15)
```

```python
x = np.linspace(-2, 10, 500)
y1 = 3/4*x-3
y2 = 2-1/7*x
plt.plot(x, y1, label='4y-3+12', c="orange")
plt.plot(x, y2, label='7y+x-14', c='blue')
plt.ylim(-3, 3)
plt.grid(True)
plt.axis('equal')
plt.legend(loc='upper left')
plt.show()
print('угол между прямыми равен ', (3/4+1/7)/(1+21/4)*180*np.pi, "градусов")
```
