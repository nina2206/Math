```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
x = np.linspace(-10, 10, 500) #создает массив из 500 значений х от -10 до 10
plt.plot(x, 1*x**3-4*x**2-1, 'red')
print(np.poly([-2.,  2.,  1.]))                                     #вычисляет коэффициенты полинома по заданным корням
plt.plot(x, 1*x**3-1*x**2-4*x+4-1, 'purple')                     #1-й способ построить полином(строит, вычисляя по формуле)
plt.plot(x, np.polyval([1., -1., -4.,  4.], x), 'orange')        #2-й способ построить полином(строит по заданным коэффициентам)
plt.xlabel('x')   #подпись оси х
plt.ylabel('y')    # подпись оси y
plt.axis('equal') #задает одинаковый масштаб по осям,  можно задать масштаб массивом по осям х и у
plt.ylim(-10, 10) # диапазон оси y 
plt.grid(True) #сетка
plt.show() #рисует график
print(np.roots([1., -1., -4.,  4.]))                #вычисляет корни полинома с заданными коэффициентами
```

```python
# графики функций из ДЗ 33.4
x = np.linspace(-4, 6, 1000)
plt.grid()
plt.axis([-5, 5, -3, 30])
plt.plot(x, np.polyval([1, 0, -1], x), label='x^2-1', c='red') 
plt.plot(x, (np.exp(x)-1)/x+1, label='(exp(x)-1)/x+1', c='purple')
plt.legend(loc='upper left')
print("система имеет решения при -2<х2 и при х>4")
plt.show()
```

```python
#преобразует полярные координаты в декартовы
R = 3
corner = np.pi/2
decX = R*np.cos(corner)
decY = R*np.sin(corner)
print(decX)
print(decY)
#преобразует декартовы координаты в полярные
x = 2
y = 3
polarR = np.sqrt(x**2+y**2)
polar_corner = math.atan(y/x)
print(polarR)
print(polar_corner)
```

```python
# график прямой в полярных координатах
x = np.linspace(0, 5, 100)
y = x+2
plt.polar(x, y)
```

```python
# график окружности в полярных координатах
x = np.linspace(0, 5, 100)
R = 5
y = np.sqrt(R**2-x**2)
plt.polar(x, y)
```

```python
# решение системы из ДЗ 33.4
x = np.linspace(-5, 5, 1000)
plt.grid()
plt.axis([-5, 5, -6, 30])
plt.plot(x, np.exp(x)-np.polyval([1, 0, -2, 1], x), 'blue')
print("Корни полинома, заданного красной линией:")
resh = np.roots([1, 0, -2, 1])
print(resh)
plt.show()
```
